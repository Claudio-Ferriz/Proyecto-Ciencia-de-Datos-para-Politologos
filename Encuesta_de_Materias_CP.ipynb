# Librerías de Python

```python
# Importación y ejecución de las librerías de Python para el análisis y modelado
# de datos.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import random
```

# Dataframes

## Dataframe Encuestas de Materias

```python

# Generación del dataframe Encuestas de Materias con datos dummy (datos ficticios)

# Generación de datos dummy

n_registros = 1560

# Columna ID Encuesta (Object/String)
ids = [f"ENC-{str(i+1).zfill(4)}" for i in range(n_registros)]

# Columna Fechas (Datetime)
base_date = datetime(2023, 1, 1)
fechas = [base_date + timedelta(days=random.randint(0, 500), minutes=random.randint(0, 1439)) for _ in range(n_registros)]

# Columna Materias
materias_list = [
    "Fundamentos de Ciencia Política I", "Teoría Política y Social I", "Economía",
    "Fundamentos de Ciencia Política II", "Teoría Política y Social II", "Historia Contemporánea",
    "Filosofía y Métodos de las Ciencias Sociales", "Historia", "Latinoamericana",
    "Política Latinoamericana", "Teoría Sociológica", "Metodología de la Investigación en Ciencia Política I",
    "Metodología de la Investigación en Ciencia Política II", "Teoría Política Contemporánea",
    "Historia Argentina", "Política Argentina", "Teoría y Derecho Constitucional",
    "Sociología Política", "Filosofía", "Teoría de las Relaciones Internacionales",
    "Opinión Pública", "Sistemas Políticos Comparados", "Administración y Políticas Públicas"
]
materias = [random.choice(materias_list) for _ in range(n_registros)]

# 4. Columna Dictado
dictado_opciones = ["presencial", "hibrido", "virtual"]
dictados = [random.choice(dictado_opciones) for _ in range(n_registros)]

# Generación del DataFrame
df_encuestas = pd.DataFrame({
    'encuesta': ids,
    'fecha de encuesta': fechas,
    'materia': materias,
    'catedra': [None] * n_registros,
    'comisión': [random.randint(1, 8) for _ in range(n_registros)],
    'dictado de la materia': dictados,
    'evaluación': [random.randint(1, 10) for _ in range(n_registros)],
    'opiniones generales de la materia': [None] * n_registros
})

# Asegurar tipos de datos específicos
df_encuestas['comisión'] = df_encuestas['comisión'].astype('int64')
df_encuestas['evaluación'] = df_encuestas['evaluación'].astype('int64')

```
